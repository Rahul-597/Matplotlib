```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('dark_background')
```

```python
clasess = ['Physics','Chemistry','Maths','Science','SST']
marks = [89,45,78,23,90]
```

```python
plt.pie(marks,labels = clasess)
plt.show()
```

```python
colors = ['red','blue','green','orange','white']
plt.pie(marks,labels = clasess,colors = colors)
plt.show()
```

```python
colors = ['red','blue','green','orange','white']
plt.pie(marks,labels = clasess,colors = colors,autopct = '%0.2f%%')
plt.show()
```

```python
exploded_values = [0.1,0,0,0.2,0]
plt.pie(marks,labels = clasess,colors = colors,autopct = '%0.1f%%',explode = exploded_values)
plt.show()
```

```python
exploded_values = [0.1,0,0,0.2,0]
plt.pie(marks,labels = clasess,colors = colors,autopct = '%0.1f%%',explode = exploded_values,shadow = True)
plt.show()
```

```python
textprops = {'fontsize':14,'color':'r'}
plt.pie(marks,labels = clasess,colors = colors,autopct = '%0.1f%%',explode = exploded_values,shadow = True,radius = 1.6,textprops = textprops)
plt.show()
```

```python
textprops = {'fontsize':14,'color':'r'}
wedgeprops = {'linewidth':3,'linestyle':'--','edgecolor':'white'}
plt.pie(marks,labels = clasess,colors = colors,autopct = '%0.1f%%',explode = exploded_values,shadow = True,radius = 1.6,textprops = textprops,wedgeprops = wedgeprops)
plt.show()
```
