```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('dark_background')
```

```python
subjects = ['Maths','English','Science','Social Studies','Computer']
marks = [89,90,45,78,99]
```

```python
plt.bar(subjects,marks)
plt.show()
```

```python
plt.bar(subjects,marks,color = 'green')
plt.show()
```

```python
colors = ['red','blue','green','orange','purple']
plt.bar(subjects,marks,color = colors)
plt.show()
```

```python
plt.bar(subjects,marks,color = colors,width = 0.5)
plt.show()
```

```python
plt.bar(subjects,marks,color = colors,edgecolor = 'white')
plt.show()
```

```python
plt.bar(subjects,marks,color = colors,edgecolor = 'white',linewidth = 4)
plt.show()
```

```python
plt.bar(subjects,marks,color = colors,edgecolor = 'white',linewidth = 4,linestyle = '-.')
plt.show()
```

```python
plt.barh(subjects,marks,color = colors)
plt.show()
```

```python
subjects = ['Maths','English','Science','Social Studies','Computer']
marks1 = [89,90,45,78,99]
marks2 = [75,94,86,80,99]
```

```python
plt.figure(figsize = (8,8))
plt.bar(subjects,marks1)
plt.bar(subjects,marks2)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
subject_len = np.arange(len(subjects))
width = 0.4
```

```python
plt.figure(figsize = (8,8))
plt.bar(subject_len,marks1,width = width)
plt.bar(subject_len+width ,marks2,width = width)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.bar(subject_len,marks1,width = width,color = colors)
plt.bar(subject_len+width ,marks2,width = width,color = colors,alpha = 0.5)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
